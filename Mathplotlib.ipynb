### Pylot API và Object-Oriented API

#### Pylot : quickly
#### OO : advanced

```python
x = np.linspace(0,10,1000)
x
```

```python
plt.plot(x,np.sin(x),color = "green",label="sin(x)")
plt.plot(x,np.cos(x),color = "blue",linestyle = "dashed",label="cos(x)")
plt.legend() #chú thích label
plt.xlabel("Biến x")
plt.ylabel("F(x)")

#tên
plt.title("Sin(x) and Cos(x)")

plt.xlim([0,5])
#plt.ylim([-0.75,0.5])
#HOẶC plt.axis([xmin,xmax,ymin,ymax])
```

```python
plt.plot(x,np.sin(x),color = "g",label="sin(x)")
plt.plot(x,np.cos(x),color = "blue",linestyle = "-",label="cos(x)")
plt.legend() #chú thích label
plt.xlabel("Biến x")
plt.ylabel("F(x)")


#tên
plt.title("Sin(x) and Cos(x)")

plt.axis("equal")
```

```python

plt.plot(x,np.sin(x),color = "green",label="sin(x)")
plt.plot(x,np.cos(x),'b--',label="cos(x)")
plt.legend() #chú thích label
plt.xlabel("Biến x")
plt.ylabel("F(x)")


#tên
plt.title("Sin(x) and Cos(x)")
plt.axis("tight")
```

```python
#OO API

# prepare
x = np.random.randn(100) #100 x ngẫu nhiên: mean=0, độ lệch chuẩn = 1
y = np.random.randn(100)

sizes = 500*np.random.rand(100) #kích cỡ tùm lum
colors = np.random.rand(100)

fig,ax = plt.subplots()
img1 = ax.scatter(x,y,s = sizes,c=colors, cmap='viridis', alpha = 0.5); #cmap: màu, alpha: độ trong suốt

fig.colorbar(img1); #màu tương ứng mức độ
```

## Biểu diễn PHÂN PHỐI của một tập dữ liệu số

```python
#prepare data

stu_height = np.random.normal(170,10,250)
stu_height[:10]
```

```python
#PYPLOT API
plt.hist(stu_height);
```

```python
#cụ thể hơn: bins là số lượng khoảng trong biểu đồ(số cột)
plt.hist(stu_height,bins = 30,alpha = 0.7);
```

```python
# OPTION 1:
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize=(10,5))

# Plot data to each axis
ax1.plot(x,x/2);
ax1.plot(x,x*2);

ax2.scatter(np.random.random(10),np.random.random(10));

ax3.bar(drinks.keys(),drinks.values());

ax4.hist(stu_height,bins = 30);
```

```python
# OPTION
fig, ax = plt.subplots(nrows=2,
                      ncols=2,)

# Plot data to each axis
ax[0,0].plot(x,x/2);
ax[0,0].plot(x,x*2);

ax[0,1].scatter(np.random.random(10),np.random.random(10));

ax[1,0].bar(drinks.keys(),drinks.values());

ax[1,1].hist(stu_height,bins = 30);
```

# project: Data visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('data_california_cities.csv')
cities.head()
```

```python
# Extract Latd(vĩ độ) and Longd(kinh độ)
 
lat,lon = cities['latd'],cities['longd']

pop,area = cities['population_total'],cities['area_total_km2']
plt.style.use('seaborn')
plt.figure(figsize=(6,8))

#plot using Pylot API

plt.scatter(lon,lat,c = np.log10(pop),cmap = 'viridis',s=area,alpha = 0.5,linewidths=0)
plt.colorbar(label='log$_{10}$(population)')
plt.axis('equal')
plt.xlabel("Kinh độ")
plt.ylabel("vĩ độ");
plt.clim(3,7)

#Create a legend for cities'sizes
area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area,label=str(area)+'km$^2$',c='k',alpha = 0.5)
plt.legend(labelspacing = 1)
```

# Violin splot: kết hợp giữa biểu đồ box plot và biểu đồ mật độ xác suất

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Vẽ biểu đồ violin
plt.violinplot(data, showmeans=False, showmedians=True)

# Đặt tên trục và tiêu đề
plt.xlabel('Distribution')
plt.ylabel('Value')
plt.title('Violin Plot')
```
